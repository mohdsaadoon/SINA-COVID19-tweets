```python
#### Create Edge Table , Source are reply-tweet-id , Target are original tweet id

df_edge_table = pd.DataFrame()
df_edge_table['Source'] = reply_tweet['username']
df_edge_table['Target'] = reply_tweet['reply_to_username']
df_edge_table['reply_text'] = reply_tweet['text']
```
