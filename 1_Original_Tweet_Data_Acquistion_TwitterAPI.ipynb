```python
import os
import tweepy as tw
import pandas as pd

import csv
import datetime
```

```python
### Twitter Developer Account Credentials
consumer_key = 'xxxx'
consumer_secret = "xxx"
access_token = "xx-xx"
access_token_secret = "xxxx"
```

```python
auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tw.API(auth, wait_on_rate_limit=True)
```

```python
search_words = "covidvaccination"

filepath = "data/Hnin_Tweeter_Api/"
db_tweets = pd.DataFrame(columns= ['tweet_id', "username", "created_at",
                                   "hashtags", "text", 
                                   "reply_to_user", "is_retweeted"])
```

```python
## Collect Tweets
tweets_list = tw.Cursor(api.search, q=search_words, count= 10, lang='en').items()
```

```python
# Save the copy of an object
tweets_list_copy = tweets_list
```

```python
### File name to save original tweet
filename = filepath + 'twitter_data_analysis' + (datetime.datetime.now().strftime("%Y-%m-%d-%H")) + '.csv'
```

```python
for tweet in tweets_list:
    tweet_id = tweet.id
    created_at = tweet.created_at
    username = tweet.user.screen_name
    reply_to_user = tweet.in_reply_to_user_id
    text_encoded = tweet.text.encode('utf-8')
#   text_deecoded = text_encoded.decode('utf-8')
    is_retweeted = tweet.retweeted

    original_hashtags_list = tweet.entities['hashtags']
    trim_hashtags_list = []  
    for hashtag in original_hashtags_list:
        trim_hashtags_list.append(hashtag['text'])
    hashtags = trim_hashtags_list
    print(hashtags)
    
    #Append to dataframe - db_tweets
    data = [tweet_id, username, created_at, hashtags, text_encoded, reply_to_user, is_retweeted]
    
    with open( filename, 'a+', encoding='utf-8', newline='') as csvFile:
        csvWriter = csv.writer(csvFile)
        csvWriter.writerow(data)
```
